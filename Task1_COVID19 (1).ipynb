```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("h1n1_vaccine_prediction.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# Calculate vaccination rates by demographic group:
df.groupby('sex')['h1n1_vaccine'].mean()
df.groupby('race')['h1n1_vaccine'].mean()
# ... etc. for other demographic groups
```

```python
#bar chart of vaccination rates by sex:
df.groupby('sex')['h1n1_vaccine'].mean().plot(kind='bar')
```

```python
#Correlation between vaccination and worry about H1N1
df['h1n1_vaccine'].corr(df['h1n1_worry'])
```

```python
# Difference in vaccination rates between employed and unemployed individuals
df.groupby('employment')['h1n1_vaccine'].mean().diff()
```

```python
# Cross-tabulation of vaccination status and income level
pd.crosstab(df['h1n1_vaccine'], df['income_level'])
```

```python
#Vaccination rates for individuals who bought face masks:
df[df['bought_face_mask'] == 1]['h1n1_vaccine'].mean()
```

```python
#Demographic characteristics of unvaccinated individuals
df[df['h1n1_vaccine'] == 0].describe()
```

```python
#Histogram of worry about H1N1:
df['h1n1_worry'].hist()
```

```python
# Create a bar chart of vaccination rates by race
df.groupby('race')['h1n1_vaccine'].mean().plot(kind='bar')
```

```python
# Create a box plot of worry about H1N1
df.boxplot(column='h1n1_worry')
```

```python
# Find Number of People Vaccinated And Unvaccinated:

vaccinated_individuals = df[df["h1n1_vaccine"] == 1] 
unvaccinated_individuals = df[df["h1n1_vaccine"] == 0]

print(f"Number of vaccinated individuals: {len(vaccinated_individuals)}")
print(f"Number of unvaccinated individuals: {len(unvaccinated_individuals)}")
```
